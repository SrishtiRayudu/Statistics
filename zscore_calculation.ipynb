```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(-3, 3, 100)
df = pd.DataFrame(x)
```

```python
s = np.std(x)
m = np.mean(x)
z_score= (x - m)/ s
df['zscore'] = z_score
dd = z_score * s + m
df['derived data']= dd
```
