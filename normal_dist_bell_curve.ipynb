```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 50)
```

```python
def graph_1(x):
    s = np.std(x)
    m = np.mean(x)
    base = 1 / (s* np.sqrt(np.pi * 2)) 
    expo = np.exp(( -np.power(x - m, 2))/2*s*s)
    y=base* expo
    return y

def graph_2(x):    
    s = 1
    m = 0
    base = 1 / (np.sqrt(np.pi * 2)) 
    expo = np.exp(( -np.power(x , 2))/2)
    y=base* expo
    return y
    
```

```python
plt.plot(x, graph_1(x) , c='r')
plt.plot(x, graph_2(x) )

plt.show()
```
