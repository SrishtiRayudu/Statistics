```python
import numpy as np
import pandas as pd
from scipy.integrate import quad
```

```python
num = [i for i in np.arange(0, 3.99+ .01 , .01)]
#num
```

```python
prob =[]
def pdf_s(x):    
    s = 1
    m = 0
    base = 1 / (np.sqrt(np.pi * 2)) 
    expo = np.exp(( -np.power(x , 2))/2)
    return base* expo


prob = [[round(i,3), quad(pdf_s, np.NINF,i)[0]] for i in num]
prob
```
