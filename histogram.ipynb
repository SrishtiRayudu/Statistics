```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pop = 100
sample = 30
bin = 20

x = np.random.randint(1,10001,pop)
m = [   np.mean(np.random.choice(x, sample)) for i in range(500)    ]
h = plt.hist(m , bin)
plt.plot((np.linspace(min(m), max(m), bin)), h[0])
plt.show()
```
