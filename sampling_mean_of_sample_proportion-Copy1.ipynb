```python
import numpy as np
import collections as c
```

```python
pop_size = 1000
nomi = ['pass', 'fail']
pop = np.random.choice(nomi, 1000)
pop
```

```python
pop_count= c.Counter(pop)
pop_count['fail']/1000, pop_count['pass']/1000
```

```python
sample_size = 30
sample_dist = [np.random.choice(pop,sample_size) for i in range(100)]
sample_dist
```

```python
sample_dist_count = [c.Counter(sample_dist[i]) for i in range (100)]
sample_dist_count
```

```python
sample_dist_count_pass = [i['pass']/sample_size for i in sample_dist_count]
```

```python
sample_dist_count_fail = [i['fail']/sample_size for i in sample_dist_count]
```

```python
sum(sample_dist_count_fail)/ 100, sum(sample_dist_count_pass)/ 100
```
