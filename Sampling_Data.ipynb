## Stratified Sampling

```python
def strat(criterion,percent):
    row_len=len(tips[tips.day=='Sun'])
    v_count = int(row_len*percent/100)
    return tips.sample(v_count)
```

```python
strat('Sun',10)
```

```python
for d in np.unique(tips.day):
    
    print(d,'\n',strat(d,12),'\n')
```

## Cluster Sampling

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
m = 'Aug'
```

```python
def cluster(m):
    return flights[flights.month==m]
```

```python
cluster(m)
```

```python
dict_cluster = {}
for i in np.unique(flights.month):
    dict_cluster[i] = cluster(i)
```

```python
dict_cluster
```

```python
dict_cluster[np.random.choice(np.unique(flights.month))]
```
