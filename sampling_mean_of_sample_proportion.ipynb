```python
import numpy as np
import collections as c
```

```python
pop_size = 1000
pop = np.random.randint(1,3,pop_size)
pop
```

```python
pop_count= c.Counter(pop)
pop_count[1]/1000
```

```python
sample_size = 30
sample_dist = [np.random.choice(pop,sample_size) for i in range(100)]
sample_dist
```

```python
sample_dist_count = [c.Counter(sample_dist[i]) for i in range (100)]
sample_dist_count
```

```python
sample_dist_count_1 = [i[1]/sample_size for i in sample_dist_count]
```

```python
sample_dist_count_2 = [i[2]/sample_size for i in sample_dist_count]
```

```python
sum(sample_dist_count_1)/ 100 , sum(sample_dist_count_2)/ 100
```
